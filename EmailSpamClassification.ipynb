### Importing Required Libraries

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
# Data Set 
df = pd.read_csv("spam.csv", encoding='latin-1')
  
df.head()
```

```python
# If columns are different, fix them
df = df[['v1', 'v2']]
df.columns = ['label', 'text']
```

```python
df.head()
```

```python
# Converting labels to numbers
df['label'] = df['label'].map({'spam': 1, 'ham': 0})
```

```python
# Spliting data
X_train, X_test, y_train, y_test = train_test_split(df['text'], df['label'], test_size=0.2, random_state=42)
```

```python
# Text Preprocessing + TF-IDF vectorization
tfidf = TfidfVectorizer(stop_words='english', max_features=3000)
X_train_tfidf = tfidf.fit_transform(X_train)
X_test_tfidf = tfidf.transform(X_test)
```

# Train the model

```python
#Naive Bayes works BEST for spam classification. Using:
model = MultinomialNB()
model.fit(X_train_tfidf, y_train)
```

```python
# Making predictions
y_pred = model.predict(X_test_tfidf)
```

```python
# Evaluate
print("Accuracy:", accuracy_score(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

# Testing with a custom email

```python
sample = ["Congratulations! You won a free iPhone. Click now!"]
sample_tfidf = tfidf.transform(sample)
pred = model.predict(sample_tfidf)[0]
if pred == 1:
    print("This Email is classified as SPAM!")
else:
    print("This Email is classified as HAM-not spam")
```
